### Import the necessary libraries - pandas, numpy, seaborn, matplotlib.pyplot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Data Preprocessing

```python
# Fixing the data types
# Converting categorical columns into categorical types

CardioGoodFitness['Product'] = CardioGoodFitness.Product.astype('category')
CardioGoodFitness['Gender'] = CardioGoodFitness.Gender.astype('category')
CardioGoodFitness['MaritalStatus'] = CardioGoodFitness.MaritalStatus.astype('category')
CardioGoodFitness['Fitness'] = CardioGoodFitness.Fitness.astype('category')
```

```python
CardioGoodFitness.info()
```

Observation:
Datatypes of all columns have been fixed to reflect correct types

```python
CardioGoodFitness.describe(include='all')
```

Observation:
1. There are 3 unique products
2. Maximum age of customer is 50 and minimum age is 18
3. Maximum no. of years of education of customer is 21 and minimum is 12
